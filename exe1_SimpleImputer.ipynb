```python
from sklearn.datasets import load_iris
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
from pickle import TRUE
# Carregar conjunto de dados Iris
iris = load_iris(as_frame=TRUE)
X, y = iris.data, iris.target
```

```python
# Introduzir valores ausentes em algumas amostras
index_faltantes = [0, 5, 10]
X_dados_faltantes = X.copy()
X_dados_faltantes.loc[index_faltantes,:] = np.nan
```

```python
X_dados_faltantes.head(11)
```

```python
#visualizar valores boleano
X_dados_faltantes.isnull().sum()
```

```python
# Imputar valores ausentes usando SimpleImputer com estratégia 'mean'
#o default do missing é nan, se for "0" tem que colocar "missing_values=0"
imp_nan = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
# Aplicar o objeto imputer aos dados faltantes
dados_imputados= imp_nan.fit_transform(X_dados_faltantes)
```

```python
# Resultado
# Sintaxe: dados_faltantes[index] e dados_imputados[index] (devido ao tamanho do dataset)
dados_imputados[index_faltantes]
```
