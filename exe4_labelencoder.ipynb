Em relação ao arquivo notas.csv:
Realize o tratamento categórico usando a técnica LabelEncoder
para a coluna EVENTO MAIS RECENTE, RAZÃO SOCIAL EMITENTE e
UF_EMITENTE.
Quantas categorizações é possível obter em cada uma dessas
colunas?

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/LyndainesSantos/Machine-Learning/main/tratamento_dados/Exercicios/notas.csv")
```

```python
df.head(20)
```

```python
# Econder
encoder = LabelEncoder()
```

```python
# Categorizar
evento_recente = encoder.fit_transform(df['EVENTO MAIS RECENTE'])
```

```python
evento_recente
```

```python
# Valores únicos
np.unique(evento_recente)
```

```python
razao_social = encoder.fit_transform(df['RAZÃO SOCIAL EMITENTE'])
```

```python
razao_social
```

```python
np.unique(razao_social).shape
```

```python
uf = encoder.fit_transform(df['UF EMITENTE'])
```

```python
uf
```

```python
np.unique(uf).shape
```
